## WORD GUESSING GAME

```python
import random
```

```python
ramdom_words = ["aswin", "kavin","keerthi","vicky"]
print("choose the word ", ramdom_words)
random_names =  random.choice(ramdom_words)
for i in range(4):
    user_input = input("enter your choice").lower()
    if user_input == random_names :
        print("your guess is  currect")
        break
    elif user_input not in random_names:
        print("please select the words",ramdom_words)
    elif user_input != random_names:
        print("sorry your guess is wrong")
    else:
       pass
```
